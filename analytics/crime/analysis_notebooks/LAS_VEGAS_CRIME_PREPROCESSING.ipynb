# Feature Engineering

```python
data['Year'] = data['Event_Date'].dt.year
data['Month'] = data['Event_Date'].dt.month 
data['Month-Year'] = data['Event_Date'].dt.strftime('%B-%Y')
time_range = pd.timedelta_range(start='0 hours', end='24 hours', freq='4H')
window_labels = ['0-4', '4-8', '8-12', '12-16', '16-20', '20-24']
data['time_window'] = pd.cut(data['Event_Date'].dt.hour * 3600 + data['Event_Date'].dt.minute * 60 + data['Event_Date'].dt.second,
                      bins=time_range.total_seconds(),
                      labels=window_labels,
                      right=False)
data
```

```python
#remapping the 128 type_of_crimes into 37 type_of_crimes for better analysis
```

```python
pattern_mapping = {'Hit and Run':r'\b(?:401A)\b',
                   'Burglary':r'\b(?:406|406A|406V)\b',
                   'Robbery':r'\b(?:407|407A|407B)\b',
                   'Person with a gun/knife/other deadly weapon':r'\b(?:413|413A|413B)\b',
                   'Grand Larceny':r'\b(?:414)\b',
                   'Abuse/Neglect':r'\b(?:424)\b',
                   'Sexual Assault':r'\b(?:426)\b',
                   'Kidnap':r'\b(?:427)\b',
                   'Child Molest':r'\b(?:428)\b',
                   'Fraud':r'\b(?:432)\b',
                   'Illegal Shooting':r'\b(?:434)\b',
                   'Explosive device':r'\b(?:445)\b',
                   'Narcotics':r'\b(?: 446)\b',
                   'Accident':r'\b(?:401|401B|401C|401M)\b',
                   'Prowler':r'\b(?:403)\b',
                   'Uknown trouble/911 disconnect':r'\b(?:404|404A)\b',
                   'Suicide':r'\b(?:405)\b',
                   'Drunk driver(driving under influence)':r'\b(?:409)\b',
                   'Reckless driver':r'\b(?:410)\b',
                   'small LARCENY':r'\b(?:414A|414C)\b',
                   'Assault/Battery':r'\b(?:415|415A|415B|415C)\b',
                   'Fight/disturbance':r'\b(?:416|416B|416F)\b',
                   'Juvenile disturbance':r'\b(?:416A)\b',
                   'Family Disturbance':r'\b(?:417)\b',
                   'Suspicious':r'\b(?:425|425A|425B|425H)\b',
                   'Indecent Exposure':r'\b(?:429)\b',
                   'Animal Complaint':r'\b(?:430)\b',
                   'Stolen Property':r'\b(?:433)\b',
                   'Malicious destruction of property':r'\b(?:441)\b',
                   'Recovered Stolen Motor Vehicle':r'\b(?:411A)\b',
                   'Traffic Problem':r'\b(?:438)\b',
                   'missing person/found person/runaway':r'\b(?: 418|418A|418B)\b',
                   'Dead body':r'\b(?:419)\b',
                   'Missing/Found Property':r'\b(?:431)\b',
                   'Civil matter':r'\b(?:447)\b',
                   'Sick/injured person/mentally ill person':r'\b(?:421|422|423)\b',
                   'keep the peace':r'\b(?:437)\b',
                   'assist citizen':r'\b(?:439)\b',
                   'wanted suspect':r'\b(?:440)\b',
                   'airplane emergency':r'\b(?:442)\b',
                   'assist an officer':r'\b(?:443)\b',
                   'Officer needs help emergency/panic alarm emergency':r'\b(?:444,444A)\b'}

for category, pattern in pattern_mapping.items():
    data.loc[data['Type'].str.contains(pattern, regex=True, na=False), 'New_Type_Description'] = category

data['New_Type_Description'] = data['New_Type_Description'].fillna('Other')
data
```
