```python
import re
pattern_mapping =  {'Hit and Run':r'Traffic Collision - Hit & Run',
                   'Burglary':r'Motor Vehicle Theft|Burglary - Other|Burglary - Commercial|Burglary - Residential|Burglary - Hot Prowl',
                   'Robbery':r'Robbery - Carjacking|Robbery - Other|Robbery - Commercial|Robbery - Street|Robbery - Residential',
                   'Person with a gun/knife/other deadly weapon':r'Weapons Offense',
                   'Grand Larceny':r'Larceny Theft - Shoplifting|Larceny - Auto Parts|Larceny Theft - From Building',
                   #'Abuse/Neglect':'',
                   'Sexual Assault':r'Rape|Sex Offense|Human Trafficking, Commercial Sex Acts|Rape - Attempted',
                   'Kidnap':r'Kidnapping',
                   #'Child Molest':r'\b(?:428)\b',
                   'Fraud':r'Fraud|Forgery And Counterfeiting|Embezzlement|Bad Checks',
                   #'Illegal Shooting':r'\b(?:434)\b',
                   #'Explosive device':r'\b(?:445)\b',
                   'Narcotics':r'Drug Violation',
                   'Accident':r'Traffic Collision',
                   'Prowler':r'Stalking',
                   #'Uknown trouble/911 disconnect':r'\b(?:404|404A)\b',
                   'Suicide':r'Suicide',
                   #'Drunk driver(driving under influence)':r'\b(?:409)\b',
                   #'Reckless driver':r'\b(?:410)\b',
                   'small LARCENY':r'Larceny Theft - Other|Larceny - From Vehicle|Larceny Theft - Bicycle|Theft From Vehicle|Larceny Theft - Pickpocket|Motor Vehicle Theft (Attempted)|Larceny Theft - Purse Snatch',
                   'Assault/Battery':r'Simple Assault|Aggravated Assault|Manslaughter|Homicide|Human Trafficking',
                   'Fight/disturbance':r'Disorderly Conduct|Drunkenness',
                   #'Juvenile disturbance':r'\b(?:416A)\b',
                   #'Family Disturbance':r'\b(?:417)\b',
                   'Suspicious':r'Suspicious Occ|Suspicious Package|Loitering',
                   #'Indecent Exposure':r'\b(?:429)\b',
                   #'Animal Complaint':r'\b(?:430)\b',
                   'Stolen Property':r'Stolen Property',
                   'Malicious destruction of property':r'Vandalism|Arson',
                   'Recovered Stolen Motor Vehicle':r'Recovered Vehicle',
                   'Traffic Problem':r'Traffic Violation Arrest|Vehicle Impounded|Vehicle Misplaced',
                   'missing person/found person/runaway':r'Missing Person|Missing Adult',
                   #'Dead body':r'\b(?:419)\b',
                   'Missing/Found Property':r'Lost Property',
                   #'Civil matter':r'\b(?:447)\b',
                   #'Sick/injured person/mentally ill person':r'\b(?:421|422|423)\b',
                   #'keep the peace':r'\b(?:437)\b',
                   #'assist citizen':r'\b(?:439)\b',
                   'wanted suspect':r'Warrant',
                   #'airplane emergency':r'\b(?:442)\b',
                   #'assist an officer':r'\b(?:443)\b',
                   #'Officer needs help emergency/panic alarm emergency':r'\b(?:444,444A)\b'
                  }

for category, patterns in pattern_mapping.items():
    pattern = '|'.join(map(re.escape, patterns))
    df2.loc[df2['Incident Subcategory'].str.contains(pattern, regex=True, na=False), 'Type_Description'] = category

# Fill any missing values in the "New_Type_Description" column with 'Other'
df2['Type_Description'] = df2['Type_Description'].fillna('Other')
```

```python
df2.drop("Incident Subcategory", axis=1, inplace=True)
```

```python
df2=df2.rename(columns={"Incident Year":"Year","Intersection":"General_Location"})
df2["City"]="San Francisco"
df2
```

```python
df2['Population'] = df2['Year'].map({2018: 4726314, 2019: 4731803, 2020: 4740552, 2021: 4616610, 2022: 4579599, 2023: 4542588,2024:4505577})
df2
```

# Merging Two Datasets

```python
merged_df = pd.concat([df1, df2], ignore_index=True)
merged_df
```

```python
merged_df.isnull().sum()
```

```python
comparison_df=merged_df.groupby(["Year","City","Population"])["Type_Description"].count().reset_index(name='Crime_counts')
comparison_df=comparison_df.drop(12)
comparison_df
comparison_df.sort_values(by='Crime_counts',ascending=False)
```

```python
comparison_df['Crime_Counts_per_1000_population'] = (comparison_df['Crime_counts'] / comparison_df['Population']*1000).round(0)

# Display the DataFrame sorted by crime counts
sorted_comparison_df = comparison_df.sort_values(by='Crime_Counts_per_1000_population', ascending=False)
sorted_comparison_df
```

```python
import matplotlib.pyplot as plt
# Plotting
plt.figure(figsize=(12, 6))
for city in sorted_comparison_df['City'].unique():
    plt.plot(sorted_comparison_df[sorted_comparison_df['City'] == city]['Year'], 
             sorted_comparison_df[sorted_comparison_df['City'] == city]['Crime_Counts_per_1000_population'], 
             label=city)

plt.title('Crime Counts Over the Years')
plt.xlabel('Year')
plt.ylabel('Crime Counts')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='Crime_Counts_per_1000_population', hue='City', data=sorted_comparison_df)
plt.title('Crime Counts by City')
plt.xlabel('Year')
plt.ylabel('Crime Counts')
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.show()
```
